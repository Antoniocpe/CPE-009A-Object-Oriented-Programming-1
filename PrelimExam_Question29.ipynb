*Calculate total marks and percentage

*Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

Determine overall result (Pass/Fail)



#Performance Remarks:
Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"

```python
grade = int(input("Enter grade: "))

print("Total Marks: ", end="")
if grade >= 90 and grade <= 100:
  print("A+")
elif grade >= 80 and grade <=89:
  print("A")
elif grade >= 70 and grade <=79:
  print("B+")
elif grade >= 60 and grade <=69:
  print("B")
elif grade >= 50 and grade <=59:
  print("C")
elif grade >= 40 and grade <=49:
  print("D")
elif grade < 40:
  print("F")

print("Overall Result: ", end="")
if grade >= 40:
  print("Pass")
else:
  print("Fail")

print("Performance Remarks: ", end="")
if grade >= 90 and grade <=100:
  print("Outstanding")
elif grade >= 80 and grade <=89:
  print("Excellent")
elif grade >= 70 and grade <=79:
  print("Very Good")
elif grade >= 60 and grade <=69:
  print("Good")
elif grade >= 40 and grade <=59:
  print("Satisfactory")
elif grade <40:
  print("Needs Improvement")

```
