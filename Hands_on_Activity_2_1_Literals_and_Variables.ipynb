**Procedure:**

A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:

1.	PRELIM GRADE = 50% Prelim Exam + 50% Prelim Class Standing (CS)
2.	PRELIM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
3.	MIDTERM GRADE = 1/3 of PRELIM GRADE + 2/3 of (50% Midterm Exam + 50% Midterm Class Standing (CS))
4.	MIDTERM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
5.	FINAL GRADE = 1/3 of MIDTERM GRADE + 2/3 of (50% Final Exam + 50% Final Class Standing (CS))
6.	FINAL CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
7.	HOAs, Quizzes and Assignments are inputted as average of all submissions and are out of 100%.
8.	Major exams are inputted out of 100%.
9.	Show the codes that successfully run the program.
10.	Provide comments or documentation strings for your program.


```python
#prelim input = PG
  #PE
  # PCS = 50% PHOA + 30% PQ + 20% PAS
#midterm input = MG
  #ME
  #MCS = MHOA, PQ, MAS
#final input = FG
#FE
  # FCS = FHOA, FQ, FAS

#getting prelim inputs
print("Prelim ------------------------------------")
PE = float(input("Enter Prelim Exam:"))
PHOA = float(input("Enter Hands-on Activity:"))
PQ = float(input("Enter Quiz:"))
PAS = float(input("Enter Prelim Assignment:"))

#prelim class standing
PCS = ((PHOA / 100) * 50) + ((PQ / 100) * 30) + ((PAS / 100) * 20)

#prelim grade
PG = ((PE / 100) * 50) + ((PCS / 100) * 50)

#getting midterm inputs
print("Midterm ------------------------------------")
ME = float(input("Enter Midterm Exam:"))
MHOA = float(input("Enter Hands-on Activity:"))
MQ = float(input("Enter Quiz:"))
MAS = float(input("Enter Midterm Assignment:"))

#prelim class standing
MCS = ((MHOA / 100) * 50) + ((MQ / 100) * 30) + ((MAS / 100) * 20)

#midterm grade
MG = ((1/3 * PE)) + ((2/3) * ((ME + MCS)/2))

#getting final inputs
print("Final ------------------------------------ ")
FE = float(input("Enter Final Exam:"))
FHOA = float(input("Enter Hands-on Activity:"))
FQ = float(input("Enter Quiz:"))
FAS = float(input("Enter Final Assignment:"))

#final class standing
FCS = ((FHOA / 100) * 50) + ((FQ / 100) * 30) + ((FAS / 100) * 20)

#final grade
FG = ((1/3 * ME)) + ((2/3) * ((FE + FCS)/2))

print("------------------------------------------")
print("Final Grade: " + str(FG))

```
