#List Methods and the Mutable Property of Lists

```python
numberlist2 = numberlist
print(numberlist)
print(numberlist2)
```

```python
numberlist[0] = 6
print(numberlist)
print(numberlist2)
```

```python
numberlist2 = numberlist.copy()
print(numberlist2)
```

```python
numberlist[0] = 5
print(numberlist)
print(numberlist2)
```
