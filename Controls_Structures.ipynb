1. if statements

```python
a = 10
b = 11

if a < b:
  print ("a is less than b")
```

2. Elif Statements

```python
a = 12
b = 12
if b > a:
  print ("b is greater than a")
elif a == b:
  print ("a and b are equal")
```

3. Else statement

```python
a = 200
b = 33
if b > a:
  print ("b is greater than a")
elif a == b:
  print ("a and b are equal")
else:
  print ("a is greater than b")
```

4. Short Hand If Statement
- if only one statement to execute

```python
a = 200
b = 33
if a > b: print ("a is greater than b")
```

5. short hand if else statement

```python
a = 2
b = 330
print ("A") if a > b else print ("B")
```

nested if

```python
x = int(input("enter a number: "))


if x > 10:
  print ("above ten,")
  if x > 20:
    print ("and also above 20!")
    if x > 30:
      print ("and also above 30!")
      if x > 40:
        print ("and also above 40!")
        if x > 50:
          print ("and also above 50!")
        else:
          print ("but not above 50.")
      else:
        print ("but not above 40.")
    else:
      print ("but not above 30.")
  else:
    print ("but not above 20.")
else:
  print ("but not above 10.")

#+5
```

Example 3:

-write a program to determine if the grades are:

if:
grade >= 75, "passed"

grade = 74, "remedial"

grade < 74, "Failed"

note: if grade is a negative number, it will not be accepted

```python
while True:
  grade = float(input("enter your grade: "))

  if grade >= 75:
    print ("passed")
  elif grade == 74:
    print ("remedial")
  elif grade < 0:
    print ("invalid input")
  else:
    print ("failed")

#+5
```
