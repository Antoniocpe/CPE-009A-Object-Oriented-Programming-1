```python
#example 1
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
```

```python
#example 2

week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
  if x == "Wednesday":
    break
```

```python
#example 3

week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Wednesday":
    break
  print(x)
```

the range() function

```python
for x in range(6):
  print(x)
for x in range(2,6):
  print(x)
```

Nested loops

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x,y)
```

The break statement

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

the continue statemnt

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print (i)
```
