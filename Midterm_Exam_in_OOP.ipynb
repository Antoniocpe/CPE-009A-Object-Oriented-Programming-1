```python
#Question 24

#Write a function that checks if a string of parentheses (), brackets [], and braces {} is balanced.

string = ["(())", "({[]})", "({)}"]

start = ["(", "[", "{"]
end = [")", "]", "}"]

i = 0

def check(string):
  for i in range(len(string)):
    for char in string[i]:
      if char[0] in start and char[len(string[i])] in end:
        return True

for i in range(len(string)):
  print(check(string))
```
